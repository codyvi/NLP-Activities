```python
from transformers import pipeline
```

```python
summarizer = pipeline("summarization")
```

```python
import pandas as pd
df_twitter =pd.read_csv('dataTwitter(2).csv', lineterminator='\n')
df_twitter.head()
```

```python
texto_resumido_twitter=[]
cnt = 0
for text in df_twitter['text']:
  cnt += 1
  print(cnt)
  helper=summarizer(text, max_length=130, min_length=30, do_sample=False)
  texto_resumido_twitter.append(helper[0]['summary_text'])
  helper='Null'
```

```python
import csv
row_list = [["text", "sentiment"]]
for x in texto_resumido_twitter:
  row_list.append([x, 0])
with open('dataTwitterRsm(2).csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(row_list)
```
